```python
import matplotlib.pyplot as plt
import numpy as np

sample = np.random.normal(loc=50, scale=5, size=1000)

plt.figure(figsize=(5,4))
plt.hist(sample, bins=10)
plt.show()

sample_mean = np.mean(sample)
sample_std = np.std(sample)
print('Mean=%.3f \nStandard Deviation=%.3f' % (sample_mean, sample_std))
```

Percobaan 5


```python
from matplotlib import pyplot as plt
from numpy.random import normal
from numpy import mean, std
from scipy.stats import norm

# Generate a normal distribution sample
sample = normal(size=1000, loc=50, scale=5)

# Plot the histogram
plt.title('HISTOGRAM')
plt.hist(sample, bins=10, density=True)

# Calculate mean and standard deviation for the normal distribution
mu, sigma = mean(sample), std(sample)

# Create a normal distribution object
dist = norm(mu, sigma)

# Generate values for the x-axis
values = [value for value in range(30, 70)]

# Calculate the probability density function (PDF) for each value
probabilities = [dist.pdf(value) for value in values]

# Plot the PDF
plt.plot(values, probabilities, 'r-', label='PDF')

# Show the plot
plt.legend()
plt.show()
```
